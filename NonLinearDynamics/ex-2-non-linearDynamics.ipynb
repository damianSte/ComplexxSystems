```python
def midpoint_method(fx, fy, x0, y0, dt, steps):
    x = np.zeros(steps)
    y = np.zeros(steps)
    x[0], y[0] = x0, y0
    for n in range(steps - 1):
        kx = dt * fx(x[n], y[n])
        ky = dt * fy(x[n], y[n])
        
        # Midpoint evaluation
        x_mid = x[n] + kx / 2
        y_mid = y[n] + ky / 2
        
        x[n + 1] = x[n] + dt * fx(x_mid, y_mid)
        y[n + 1] = y[n] + dt * fy(x_mid, y_mid)
    return x, y
```

```python
# Equation 1: x'' + x = 0
def fx1(x, y):
    return y

def fy1(x, y):
    return -x

# Equation 2: x'' + sin(x) = 0
def fx2(x, y):
    return y

def fy2(x, y):
    return -np.sin(x)

# Equation 3: x'' = -x + x^3
def fx3(x, y):
    return y

def fy3(x, y):
    return -x + x**3

# Equation 4: x'' = x - x^3
def fx4(x, y):
    return y

def fy4(x, y):
    return x - x**3
```

```python
def plot_phase_portrait(fx, fy, title, xlim=(-3, 3), ylim=(-3, 3), num_trajectories=20):
    # Create a grid of initial conditions
    x0_vals = np.linspace(xlim[0], xlim[1], num_trajectories)
    y0_vals = np.linspace(ylim[0], ylim[1], num_trajectories)
    
    plt.figure(figsize=(8, 6))
    
    for x0 in x0_vals:
        for y0 in y0_vals:
            x, y = midpoint_method(fx, fy, x0, y0, dt=0.05, steps=100)
            plt.plot(x, y, color='blue', alpha=0.3)
            
            x_values = np.linspace(xlim[0], xlim[1], 100)
             
            for s in np.linspace(0, 2, 10):
                  y_values = s * x_values
                  plt.plot(x_values, y_values, 'r--')
    
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.grid(True)
    plt.show()
```
