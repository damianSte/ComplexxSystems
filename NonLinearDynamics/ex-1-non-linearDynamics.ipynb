```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
def f(x):
    return x*(x-1)*(x-2)
```

```python
fixed_points = np.roots ([1,-3,2,0])
print("Fixed points:" ,fixed_points)
```

```python
def euler_method(x0, dt, t_max):
    t = np.arange(0, t_max, dt)
    x = np.zeros(len(t))
    x[0] = x0

    for n in range(1, len(t)):
        x[n] = x[n - 1] + dt * f(x[n - 1])

    return t, x
```

```python
t_max = 10
initial_conditions = [-0.1,0.1, 1.1, 1.9, 2.1]
time_steps = [0.1, 0.01, 0.001]
```

```python
plt.figure(figsize=(12,6))
for dt in time_steps:
    for x0 in initial_conditions:
        t_values, x_values = euler_method(x0, dt, t_max)
        plt.plot(t_values, x_values, label=f'x0={x0}, dt={dt}')

plt.title('Euler Method: Stability of Fixed Points for x(x-1)(x-2)')
plt.xlabel('Time (t)')
plt.ylabel('x(t)')
plt.ylim(-2,4)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```
